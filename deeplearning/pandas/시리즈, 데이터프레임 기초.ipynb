```python
import numpy as np
import pandas as pd
```

```python
print("시리즈는 1차원 레이블이 붙은 배열, 레이블을 인덱스라고 한다.")
s = pd.Series([1, 3, 5, np.nan, 6, 8])

print(s)

print("\n인덱스를 지정해서 생성할 수도 있다.")
s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
print(s)

print(s.index)

print("\n딕셔너리에서 시리즈 생성")
d = {'b': 1, 'a': 0, 'c': 2}
print(pd.Series(d))

print("\n인덱스를 입력하면 인덱스에 일치하는 데이터 순서로 구성됨, 없으면 Nan")
print(pd.Series(d, index=['b', 'c', 'd', 'a']))

print("\n스칼라 값에서 생성하면 값이 반복된다.")
print(pd.Series(5., index=['a', 'b', 'c', 'd']))

print("\n시리즈는 numpy배열과 매우 유사, 슬라이싱은 인덱스도 같이 슬라이싱 함")
print(s[0])

print(f"\n{s[:3]}")
print(f"\n{s.median()}\n{s[s > s.median()]}")
print(f"\n{s[[3, 1, 0]]}")
print(f"\n{np.exp(s)}")




# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
# print(f"\n{}")
```

```python
print("\npandas배열이 필요하다면(인덱스 없이 연산하는 경우)")
print(s.array)

print("\nnumpy배열이 필요하다면")
print(s.to_numpy())

print("\n시리즈는 딕셔너리와 유사하다")
print(s['a'])
s['e'] = 12
print(s)
print('e' in s)

print("\n벡터 계산이 가능하며, 결과는 인덱스에 따라 정렬")
print(s + s)
print(s * 2)
print(s[1:] + s[:-1])

print("\n시리즈에 이름 붙이기")
s2 = pd.Series(np.random.randn(5), name="어쩌고")
print(s2)
s3 = s2.rename("저쩌고")
print(s3.name)
```

```python
#데이터 프레임은, 2차원 인덱싱된 데이터 구조. SQL테이블이나 시리즈 객체의 딕셔너리로 생각해도 됨.

print("\n시리즈의 딕셔너리 목록이나 딕셔너리 목록에서 생성")
d = {
    'one': pd.Series([1., 2., 3.], index=['a', 'b', 'c']),
    'two': pd.Series([1., 2., 3., 4.], index=['a', 'b', 'c', 'd'])
}
df = pd.DataFrame(d)
print(df)
print(f"\n{pd.DataFrame(d, index=['d', 'b', 'a'])}")
print(f"\n{pd.DataFrame(d, index=['d', 'b', 'a'], columns=['two', 'three'])}")
print(f"\n{df.index}")
print(f"\n{df.columns}")

print("\nnumpy 배열이나 리스트에서 생성, index가 없다면 range(n)으로 붙음")
d = {
    'one': [1., 2., 3., 4.],
    'two': [4., 3., 2., 1.]
}
print(pd.DataFrame(d))
print(pd.DataFrame(d, index=['a', 'b', 'c', 'd']))

print("\nstructures 배열")
data = np.zeros((2,), dtype=[('A', 'i4'), ('B', 'f4'), ('C', 'a10')])
data[:] = [(1, 2., 'Hello'), (2, 3., "World")]

print(pd.DataFrame(data))
print(pd.DataFrame(data, index=['first', 'second']))
print(pd.DataFrame(data, columns=['C', 'A', 'B']))

print("\n딕셔너리 목록에서")
data2 = [
    {'a': 1, 'b': 2},
    {'a': 5, 'b': 10, 'c': 20}
]
print(pd.DataFrame(data2))
print(pd.DataFrame(data2, index=['first', 'second']))
print(pd.DataFrame(data2, columns=['a', 'b']))

print("\n튜플로 구성된 딕셔너리 목록에서(멀티 인덱스 프레임을 구성)")
e = pd.DataFrame(
    {
        ('a', 'b') : {('A', 'B'): 1, ('A', 'C'): 2},
        ('a', 'a') : {('A', 'C'): 3, ('A', 'B'): 4},
        ('a', 'c') : {('A', 'B'): 5, ('A', 'C'): 6},
        ('b', 'a') : {('A', 'C'): 7, ('A', 'B'): 8},
        ('b', 'b') : {('A', 'D'): 9, ('A', 'B'): 10}
    }
)
print(e)
```

```python
print("데이터 프레임을 생성하는 다른 방법(딕셔너리)")
f = dict([
    ('A', [1, 2, 3]),
    ('B', [4, 5, 6])
])

print(pd.DataFrame(f))
print(pd.DataFrame.from_dict(f))
#orient의 기본값은 columns
print(pd.DataFrame.from_dict(f, orient="index", columns=['one', 'two', 'three']))

print("\n튜플 목록이나 structured numpy배열에서 생성하기")
data = np.zeros((2,), dtype=[('A', 'i4'), ('B', 'f4'), ('C', 'a10')])
data[:] = [(1, 2., 'Hello'), (2, 3., "World")]
print(pd.DataFrame(data))
print(pd.DataFrame.from_records(data))
#C열을 인덱스로 지정
g = pd.DataFrame.from_records(data, index='C')
print(g)
```

```python
print("열 선택, 더하기, 빼기")
d = {
    'one': pd.Series([1., 2., 3.], index=['a', 'b', 'c']),
    'two': pd.Series([1., 2., 3., 4.], index=['a', 'b', 'c', 'd'])
}
df = pd.DataFrame(d)

print(df)
print(f"\n{df['one']}")
df['three'] = df['one'] * df['two']
df['flag'] = df['one'] > 2
print(f"\n{df}")

print("\n열 제거하기")
del df['two']
three = df.pop('three')
print(df)

print("\n열 삽입하기")
df['foo'] = 'bar'
print(df)

print("\n삽입되는 값이 모자라도 데이터 형식에 맞춤")
df["one_trunc"] = df['one'][:2]
print(df)

print("\nnumpy 배열은 추가시 인덱스가 맞아야 하며, insert를 이용하면 원하는 열에 추가가능")
df.insert(1, "bar", df['one'])
print(df)
```

```python
iris = pd.read_csv('iris.data')
print(iris.head())

iris2 = iris.assign(sepal_ratio=iris['SepalWidth'] / iris['SepalLength'])
print(f"\n{iris2.head()}")

#람다 함수를 넘겨주는 것도 가능
iris3 = iris.assign(sepal_ratio=lambda x: (x['SepalWidth'] / x['SepalLength']))
print(f"\n{iris3.head()}")

iris.query('SepalLength > 5')\
    .assign(SepalRatio=lambda x: x.SepalWidth / x.SepalLength,
           PetalRatio=lambda x: x.PetalWidth / x.PetalLength)\
    .plot(kind='scatter', x='SepalRatio', y = 'PetalRatio')
```

```python
print("\nassign안에서 새로 만들어진 컬럼을 참조 가능(python 3.6, pandas 0.23.0 이상)")
dfa = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6]
})
dfa2 = dfa.assign(C=lambda x: x['A'] + x['B'],
          D=lambda x: x['A'] + x['C'])#아직 존재하지 않지만, assign에서 정의된 C를 참조
print(dfa2)
```

```python
print("\n데이터 정렬 및 연산")
df2 = pd.DataFrame(np.random.randn(10, 4), columns=['A', 'B', 'C', 'D'])
print(df2)
df3 = pd.DataFrame(np.random.randn(7, 3), columns=['A', 'B', 'C'])
print(f"\n{df3}")

print(f"\n{df2 + df3}")

print("\n데이터 프레임과 시리즈간의 연산에서는 시리즈 인덱스를 데이터프레임의 열에 맞추고 행을 브로드캐스트 하여 연산한다.")
print(df2.iloc[0])
print(f"\n{df2 - df2.iloc[0]}")
```

```python
print("\n시계열 데이터에서, 데이터프레임의 인덱스가 날짜인 경우 열 방향 브로드캐스트로 연산된다.")
index = pd.date_range('20200101', periods=8)
df4 = pd.DataFrame(np.random.randn(8, 3), index=index, columns=list('ABC'))
print(df4)

print(f"\n{df4['A']}")

print(f"\n{df4 - df4['A']}")
```

```python
df1 = pd.DataFrame({'a': [1, 0, 1], 'b': [0, 1, 1]}, dtype=bool)
df2 = pd.DataFrame({'a': [0, 1, 1], 'b': [1, 1, 0]}, dtype=bool)

print(df1)
print(df2)

print("\n", df1 & df2)
print("\n", df1 ^ df2)
print("\n", -df1)
```
